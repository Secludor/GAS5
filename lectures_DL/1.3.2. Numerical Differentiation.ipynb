```python
def numerical_diff(f, x):
    h = 0.00000001
    return (f(x+h)-f(x))/h
```

```python
def function_1(x):
    return 0.01*x**2 + 0.2*x
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0.0, 20.0, 0.1)
y = function_1(x)

plt.xlabel('x')
plt.ylabel('f(x)')
plt.plot(x,y)
plt.show()
```

```python
numerical_diff(function_1, 5), numerical_diff(function_1, 10)
```

### 편미분

```python
def function_2(x):
    return x[0]**2 + x[1]**2
```

```python
def function_2_partial1(x0):
    return function_2([x0,4.0])

def function_2_partial2(x1):
    return function_2([3.0,x1])
```

```python
numerical_diff(function_2_partial1,3.0)
```

```python
numerical_diff(function_2_partial2,4.0)
```
