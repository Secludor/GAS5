### 선 색상 수정하기
- color 속성을 이용하여 선의 색을 바꿀 수 있다
- https://matplotlib.org/3.1.0/gallery/color/named_colors.html
(matplotlib color)




```python
def my_cal(num):
    num2 = num ** 2
    return num2
x = [1, 3, 5, 7, 9]
y = [2, 4, 3, 6, 10]
y2 = []
for x_each in x: # 함수로 y2 변수 생성
    y2.append(my_cal(x_each))
plt.plot(x, y, color='lightgreen')
plt.plot(x, y2, color='darkcyan')
plt.show()
```

### 선 스타일 수정하기
선의 굵기는 linewidth=3 등으로 변경할 수 있다

- 선 스타일
- https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html (matplotlib linestyle)

```python
import matplotlib.pyplot as plt
plt.style.use("seaborn-v0_8")
def my_cal(num, i):
    num2 = num ** i
    return num2
x = [1, 3, 5, 7, 9]
y = [2, 4, 3, 6, 10]
y2 = [] ; y3 = [] # 다음 줄에 쓸 것을 세미콜론으로 엮을 수 있다
for x_each in x:
    y2.append(my_cal(x_each, 2))
    y3.append(my_cal(x_each, 3))
plt.plot(x, y, color='red', linestyle='dashed')
plt.plot(x, y2, color='darkcyan', linestyle='dashdot')
plt.plot(x, y3, color='forestgreen', linestyle='dotted')
plt.show()
```

## 상자 그래프 boxplot
- 기본 통계에 대한 요약 정보를 얻을 수 있음.(다섯 수치의 요약)
    - ※ 다섯수치요약: 사분위수(1/4 , 2/4 , 3/4 , 4/4)와 최소값을 말한다

- 높은 이상치 : 제3사분위수 + 1.5 * IQR 보다 높은 값
- ※ IQR(사분위편차) = 제3사분위수 – 제1사분위수

## 상관관계 함수 corr()
- 변수 사이의 상관관계는 상관관계 함수로 살펴볼 수 있다.

- 옵션
    - numueric_only: 숫자형 변수만 적용할지 여부 (기본값 False)

※ corr() 함수의 method 파라미터
(기본값은 pearson)
- pearson : standard correlation coefficient
- kendall : Kendall Tau correlation coefficient
- spearman : Spearman rank correlation

```python
# 피어슨 상관계수
print(df.Temperature.corr(df.Humidity))
print(df.Humidity.corr(df.CO2))
print(df.HumidityRatio.corr(df.Humidity))
```

```python
#스피어만 상관계수
print(df.Temperature.corr(df.Humidity, method="spearman"))
print(df.Humidity.corr(df.CO2, method="spearman"))
print(df.HumidityRatio.corr(df.Humidity, method="spearman"))
```

```python
df.corr(numeric_only=True, method='pearson')
```

```python
df.corr(numeric_only=True, method='kendall')
```

```python
df.corr(numeric_only=True, method='spearman')
```

0.7 이상이면 강한 상관관계가 있다.

이 경우 몇 가지 해결 방안이 있다.
1. plt.tight_layout()을 이용해 간격을 자동으로 조정한다.

```python
plt.figure(figsize=(7, 5))
plt.subplot(211) # 2x1행렬에 넣는다
plt.plot(df.Temperature, df.CO2, marker=".", linestyle="")
plt.title("Temperature & CO2")
plt.subplot(223) # 2x2행렬의 3번째 칸에 넣는다
plt.hist(df.HumidityRatio)
plt.title("HumidityRatio")
plt.subplot(224, frameon=False) # 2x2행렬의 4번째 칸에 넣는다
plt.boxplot(df.Humidity)
plt.title("Humidity")
plt.tight_layout() # subplot사이의 간격 자동 조정
plt.show()
```

2. 간격을 수동으로 조정할 수 있는 subplots_adjust() 함수를 사용한다.

```python
plt.figure(figsize=(7, 5))
plt.subplot(211) # 2x1 행렬에 넣는다
plt.plot(df.Temperature, df.CO2, marker=".", linestyle="")
plt.title("Temperature & CO2")
plt.subplot(223) # 2x2 행렬의 3번째 칸에 넣는다
plt.hist(df.HumidityRatio)
plt.title("HumidityRatio")
plt.subplot(224, frameon=False) # 2x2 행렬의 4번째 칸에 넣는다
plt.boxplot(df.Humidity)
plt.title("Humidity")
plt.subplots_adjust(wspace=1, hspace=1) # subplot 사이의 간격 수동 조정
plt.show()
```
