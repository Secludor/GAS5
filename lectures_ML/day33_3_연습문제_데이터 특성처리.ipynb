```python
from sklearn.metrics import mean_squared_error

def train_and_evaluate(X_train, X_test, y_train, y_test):
    model = LinearRegression()
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    
    print(model.score(X_train, y_train))
    print(model.score(X_test, y_test))
    
    return mse
```

```python
mse_orginal = train_and_evaluate(X_train, X_test, y_train, y_test)
mse_mms = train_and_evaluate(X_train_mms, X_test_mms, y_train, y_test)
mse_ss = train_and_evaluate(X_train_ss, X_test_ss, y_train, y_test)

print(mse_orginal, mse_mms, mse_ss)
```

```python
data = pd.read_csv('BostonHousing.csv')
print(data.head())
print(data.shape)

X = data.iloc[:,0:13]
y = data.iloc[:,13]

print(X.shape)
print(y.shape)
print(X.head())
print(y.head())

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

from sklearn.linear_model import LinearRegression
model_L = LinearRegression()
model_L.fit(X_train,y_train)

print('원본 테스트 데이터 점수: ',model_L.score(X_test,y_test))

from sklearn.preprocessing import MinMaxScaler
scaler1 = MinMaxScaler()
scaler1.fit(X_train)
X_train_scaled1 = scaler1.transform(X_train)
X_test_scaled1 = scaler1.transform(X_test)

from sklearn.preprocessing import StandardScaler
scaler2 = StandardScaler()
scaler2.fit(X_train)
X_train_scaled2 = scaler2.transform(X_train)
X_test_scaled2 = scaler2.transform(X_test)

model1 = LinearRegression()
model1.fit(X_train_scaled1,y_train)
print('MinMaxScaler 테스트 데이터 점수: ',model1.score(X_test_scaled1,y_test))

model2 = LinearRegression()
model2.fit(X_train_scaled2,y_train)
print('StandardScaler 테스트 데이터 점수: ',model2.score(X_test_scaled2,y_test))
```
