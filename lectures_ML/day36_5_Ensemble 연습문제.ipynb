### 연습문제 2
- 각 모델들을 회귀분석에 사용해보기.

- reg.csv 파일을 대상으로 다음 알고리즘으로 회귀분석을 수행하시오.
    - Bagging(DecisionTree) 이용
    - RandomForest
    - Extra-Trees
    - Adaboost
    - GradientBoosting
    - HistGradientBoosting
    - XGBoost
    - LightGBM

```python
import os
import pandas as pd

path = r'D:\elice_python\GAS_5\pytest_machine'

os.chdir(path)
```

```python
data = pd.read_csv('reg.csv', header=0)
print(data.head(1))
print(data.shape)
```

```python
X = data.iloc[:,:-1]
y = data.iloc[:,-1]
print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import (
    BaggingRegressor, 
    RandomForestRegressor,
    ExtraTreesRegressor,
    AdaBoostRegressor,
    GradientBoostingRegressor,
    HistGradientBoostingRegressor
                              )
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
```

```python
bag = BaggingRegressor(estimator=DecisionTreeRegressor(), n_estimators=100, random_state=42)
randf = RandomForestRegressor(n_estimators=100, random_state=42)
extree = ExtraTreesRegressor(n_estimators=100, random_state=42)
adab = AdaBoostRegressor(n_estimators=100, random_state=42)
gb = GradientBoostingRegressor(n_estimators=100, max_depth=3, random_state=42)
hgb = HistGradientBoostingRegressor()

xgb = XGBRegressor(n_esimators=100)
lgb = LGBMRegressor(n_estimators=100)


for i, model in enumerate([bag, randf, extree, adab, gb, hgb, xgb, lgb]):
    model.fit(X_train, y_train)
    print(i, model.score(X_train,y_train))
    print(i, model.score(X_test,y_test))
    print()
```

### 연습문제 3
- 산탄데르 고객 만족 데이터 세트(Santander Customer Satisfaction)는 370개 독립변수로 이루어진 고객 만족 여부 예측용 데이터 세트이다.
- 산탄데르 은행이 캐글 경연을 의뢰한 데이터로써, 독립변수의 이름은 모두 익명 처리되어 어떤 속성인지는 알 수 없다.
- 다만 TARGET 컬럼이 종속변수이며, 값이 1이면 불만을 가진 고객, 0이면 만족한 고객을 의미한다.
- `santander_train.csv` 파일을 `random_state=1111`로 `train_test_split()`하여, RandomForest, HistGradientBoosting, XGBoost, LightBGM을 사용해 속도와 정확도를 계산하시오.

```python
data = pd.read_csv('santander/santander_train.csv', header=0)
data.head(1)
```

```python
data.info()
```

```python
X = data.iloc[:,:-1]
y= data.iloc[:,-1]

print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=1111)
```

```python
from sklearn.ensemble import RandomForestClassifier, HistGradientBoostingClassifier, ExtraTreesClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
```

```python
rdfc = RandomForestClassifier(n_estimators=50, max_features='sqrt', random_state=42)
rdfc.fit(X_train, y_train)
```

```python
print(rdfc.score(X_train, y_train))
print(rdfc.score(X_test, y_test))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=1111)
```

```python
histgb = HistGradientBoostingClassifier(random_state=42)
histgb.fit(X_train,y_train)
```

```python
print(histgb.score(X_train, y_train))
print(histgb.score(X_test, y_test))
```
