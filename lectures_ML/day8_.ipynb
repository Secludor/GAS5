```python
import matplotlib.pyplot as plt

class Point:
    def __init__(self,x,y) -> None:
        self.x = x
        self.y = y
    
    def draw(self):
        plt.plot(self.x,self.y, marker='o', markersize =10, color='blue')
```

```python
p1 = Point(1,5)
p2 = Point(-1,-5)
```

```python
print(p1.x,p1.y)
print(p2.x,p2.y)
```

```python
p1.draw()
p2.draw()
```

```python

p2.draw()
```

```python
class Circle1(Point):
    def __init__(self, x, y, r) -> None:
        super().__init__(x, y)
        self.r=r
```

```python
c1 = Circle1(1,-1,5)
print(c1.x,c1.y,c1.r)
```

```python
c1.draw()
```

```python
plt.xlim(-5,5)
plt.ylim(-5,5)
c1.draw()
plt.show()
```

```python
import matplotlib.patches as patches

ax = plt.subplot()

class Circle2(Point):
    def __init__(self, x, y, r) -> None:
        super().__init__(x, y)
        self.r = r
        
    def draw(self):
        c = patches.Circle(xy = (self.x,self.y), radius=self.r, facecolor='b', edgecolor='r')
        ax.add_patch(c) # facecolor : 채우기, edgecolor : 테두리
```

```python
c2 = Circle2(1,-1,2)
ax = plt.subplot()
plt.xlim(-5,5)
plt.ylim(-5,5)
c2.draw('r','b')
plt.show()
```

```python
class Archer:
    def __init__(self,name='ananimous',health=87,power=64,arrow=100) -> None:
        self.name = name
        self.health = health
        self.power = power
        self.arrow = arrow
        
    def shoot(self):
        if self.arrow>1 :
            self.arrow-=1
            print('화살발사!')
            print(f'잔량 : {self.arrow} 개')
        else:
            print('화살이 부족합니다.')
```

```python
arch = Archer('James')
```

```python
print(f'''
       이름 : {arch.name}(궁수)
기본 체력 : {arch.health}
    공격력 : {arch.power}
보유 화살 : {arch.arrow}
      ''')
```

```python
arch.shoot()
```

```python
class Dog:
    def __init__(self,breed,size,age,color) -> None:
        self.breed = breed
        self.size = size
        self.age = age
        self.color = color
        
        self.character = '중립'
    
    def Eat(self):
        print('얌얌얌')
    
    def Sleep(self):
        print('쿨쿨쿨')
    
    def Sit(self):
        print('기다리는 중')
    
    def Run(self):
        print('우다다다')
```

```python
class Mastiff(Dog):
    def __init__(self,breed='Neapolitan Mastiff', size='Large', age='5 years', color='Black') -> None:
        super().__init__(breed, size, age, color)
        self.character = '온순'
        
    def Sleep(self):
        print('마스티프가 얌전히 자고 있습니다.')
        return super().Sleep()
        
    def Run(self):
        print('마스티프가 천천히 걷고 있어요')
        return super().Run()
        
    def Hunt(self):
        print('마스티프가 먹잇감을 향해 달려갑니다.')
```

```python
mas = Mastiff()
mas.Run()
mas.Sleep()
```

```python
# 나호용
class Dog:
    def __init__(self, Age, Size, Color, type = '강아지'):
        self.Size = Size
        self.Age = Age
        self.Color = Color
        self.type = type

    def eat(self):
        print(f'{self.type}가 밥을 먹고 있습니다')

    def sleep(self):
        print(f'{self.type}가 자고 있습니다')

    def sit(self):
        print(f'{self.type}가 앉아 있습니다')

    def run(self):
        print(f'{self.type}가 뛰고 있습니다')

    def play(self):
        print(f'{self.type}가 놀고 있습니다')

        
class Mastiff(Dog):
    def __init__(self,name, Age, Size, Color, type = '마스티스', character = '온순함'):
        super(Mastiff,self).__init__(Age, Size, Color,type)
        self.name = name
        self.character = character

    def sit(self):
        print(f'{self.name}가 긴 다리로 앉아 있어요!')

    def play(self):
        print(f'{self.name} 가 친구랑 놀고 있어요~')

    def run(self):
        print(f'{self.name} is running')

class Maltese(Dog):
    def __init__(self,name, Age, Size, Color, type = '말티즈', character = '질투심'):
        super(Maltese,self).__init__(Age, Size, Color,type)
        self.name = name
        self.character = character

    def sit(self):
        print(f'{self.name}가 짧은 다리로 앉아 있어요!')

    def play(self):
        print(f'{self.name} 가 공놀이를 하고 있어요~')

    def run(self):
        print(f'{self.name} is running')

class ChowCHow(Dog):
    def __init__(self,name, Age, Size, Color, type = '차우차우', character = '예민함'):
        super(ChowCHow,self).__init__(Size, Age, Color, type)
        self.name = name
        self.character = character     

    def sit(self):
        print(f'{self.name}가 귀엽게 앉아 있어요!')

    def play(self):
        print(f'{self.name} 가 덤블링을 하고 있어요~')

    def run(self):
        print(f'{self.name} is running')

maltese = Maltese('초롱이',2, 'small', 'white')
print('저의 종류는', maltese.type, '입니다.')
print('제 이름은', maltese.name, '입니다')
maltese.sit()
print('말티즈는: '+ maltese.character)
maltese.play()
maltese.run()
```
