```python
import matplotlib.pyplot as plt

class Point:
    def __init__(self,x,y) -> None:
        self.x = x
        self.y = y
    
    def draw(self):
        plt.plot(self.x,self.y, marker='o', markersize =10, color='blue')
```

```python
p1 = Point(1,5)
p2 = Point(-1,-5)
```

```python
print(p1.x,p1.y)
print(p2.x,p2.y)
```

```python
p1.draw()
p2.draw()
```

```python

p2.draw()
```

```python
class Circle1(Point):
    def __init__(self, x, y, r) -> None:
        super().__init__(x, y)
        self.r=r
```

```python
c1 = Circle1(1,-1,5)
print(c1.x,c1.y,c1.r)
```

```python
c1.draw()
```

```python
plt.xlim(-5,5)
plt.ylim(-5,5)
c1.draw()
plt.show()
```

```python
import matplotlib.patches as patches

ax = plt.subplot()

class Circle2(Point):
    def __init__(self, x, y, r) -> None:
        super().__init__(x, y)
        self.r = r
        
    def draw(self):
        c = patches.Circle(xy = (self.x,self.y), radius=self.r, facecolor='b', edgecolor='r')
        ax.add_patch(c) # facecolor : , edgecolor : 
```

```python
c2 = Circle2(1,-1,2)
ax = plt.subplot()
plt.xlim(-5,5)
plt.ylim(-5,5)
c2.draw('r','b')
plt.show()
```

```python
class Archer:
    def __init__(self,name='ananimous',health=87,power=64,arrow=100) -> None:
        self.name = name
        self.health = health
        self.power = power
        self.arrow = arrow
        
    def shoot(self):
        if self.arrow>1 :
            self.arrow-=1
            print('!')
            print(f' : {self.arrow} ')
        else:
            print(' .')
```

```python
arch = Archer('James')
```

```python
print(f'''
        : {arch.name}()
  : {arch.health}
     : {arch.power}
  : {arch.arrow}
      ''')
```

```python
arch.shoot()
```

```python
class Dog:
    def __init__(self,breed,size,age,color) -> None:
        self.breed = breed
        self.size = size
        self.age = age
        self.color = color
        
        self.character = ''
    
    def Eat(self):
        print('')
    
    def Sleep(self):
        print('')
    
    def Sit(self):
        print(' ')
    
    def Run(self):
        print('')
```

```python
class Mastiff(Dog):
    def __init__(self,breed='Neapolitan Mastiff', size='Large', age='5 years', color='Black') -> None:
        super().__init__(breed, size, age, color)
        self.character = ''
        
    def Sleep(self):
        print('   .')
        return super().Sleep()
        
    def Run(self):
        print('   ')
        return super().Run()
        
    def Hunt(self):
        print('   .')
```

```python
mas = Mastiff()
mas.Run()
mas.Sleep()
```
