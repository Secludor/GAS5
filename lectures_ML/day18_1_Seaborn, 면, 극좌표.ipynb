## Seaborn
- Matplotlib을 기반으로 다양한 색상 테마와 기능을 추가한 시각화 패키지
- 자주 사용이 되나, Matplotlib과는 사용법이 조금 다르다.
- 기본적인 시각화 기능은 Matplotlib 패키지에 의존.
- 통계 기능은 Statsmodels 패키지에 의존.

set() : 그래프의 전반적인 스타일을 Seaborn 기본 스타일로 변경 
- 주요 옵션
	- context : 라벨 스타일 (paper, notebook, talk, poster)
	- style : 배경 스타일 (darkgrid, whitegrid, dark, white, ticks)
	- palette : 색상 스타일 ( deep, muted, pastel, bright, dark, colorblind)
	- font_scale : 폰트 크기

```python
from pandas import Series, DataFrame
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

path = 'D:\\elice_python\\GAS_5\\pytest_python\\'
```

```python
df = pd.read_csv(os.path.join(path, 'APT.csv'), encoding='euc-kr')
df.info()
```

```python
df.dropna(inplace=True)
df
```

### barplot

```python
sns.set()

sns.barplot(x=df.id, y=df.Price)
plt.show()
```

### boxplot

```python
sns.set(context='notebook', style='white', palette='pastel')
sns.boxplot(x=df.Price)
plt.xlabel('APT Price')
plt.show()
```

```python
sns.set(context='notebook', style='white', palette='pastel')
sns.boxplot(y=df.Price)
plt.xlabel('APT Price')
plt.show()
```

- matplotlib에서 기본값은 종형이지만, seaborn에서는 기본적으로 누워있다.
- matplotlib에서 눕히려면 vert=False를,
- seaborn에서 세우려면 데이터를 x= 대신 y= 으로 전달한다.

### 히스토그램과 밀도곡선
(밀도곡선 : 추정확률분포) ; kde=True (기본값) 이면 값의 분포를 히스토그램으로 표현하고 밀도함수 곡선을 그린다.

```python
sns.set(palette='dark')
sns.histplot(df.Price, color='r', kde=True)
plt.xlabel('APT Price')
plt.show()
```

### 산점도
relplot ; relational plot

```python
sns.set(context='talk')
sns.relplot(x='Price', y='FAR', data=df, col_order='g')
plt.xlabel('APT price')
plt.show()
```

#### 산점도 - 선 그래프
kind 파라미터에 'line'을 전달하면 선 그래프를 얻을 수 있다.

```python
sns.set(context='poster', style='darkgrid')
sns.relplot(x='Price', y='FAR', data=df, color='r', kind='line')
plt.xlabel('APT price')
plt.show()
```

### 회귀선과 신뢰구간
- 신뢰구간 : 전체 데이터를 볼 때 데이터가 정상치라고 판단할 수 있는 영역
- lmplot : linear model(선형회귀모델) plot
- 신뢰구간을 없애려면 ci=None

```python
sns.lmplot(x='FAR', y='Price', data=df)
plt.show()
```

### 레이더 차트

```python
data = pd.read_csv(os.path.join(path, 'sam.csv'), encoding='utf8')

data.info()
```

```python
data.head()
```

```python
col_names = data.columns[1:6].tolist()
print(col_names)
```

#### 데이터 전처리

```python
# 관우 데이터
kwanwoo = data[data.이름 == '관우'].values.tolist()
print(kwanwoo)
print(kwanwoo[0][1:6])
```

```python
# 레이터 차트에서 시작점으로 돌아갈 부분의 정보를 넣기 위해, 마지막에 첫 번째 원소의 값을 추가
kwanwoo = kwanwoo[0][1:6] + [kwanwoo[0][1]]
```

```python
kalyang = data[data.이름 == '제갈량'].values.tolist()
print(kalyang)
```

```python
kalyang = kalyang[0][1:6] +[kalyang[0][1]]
```

#### Radian 구하기
- 5 개 능력을 볼 것이므로, 원의 각도를 5 등분하여 각도를 측정하는 라디안 값으로 표현
- 각각은 원의 0%, 20% 40%, 60%, 80% 각도를 라디안 값으로 표현

```python
degree = [n/5 * 2 * np.pi for n in range(5)]
print(degree)
```

```python
degree+=degree[:1]

print(degree)
```

#### 그래프 그리기

```python
ax = plt.subplot(111, polar=True)    # (1,1)의 1번째 칸에 넣는다. //극 좌표계 방식
plt.xticks(degree[:-1], col_names, color = 'purple', size=10)    # 기준 정보 출력 //[:-1]로 마지막에 원점으로 돌아가기 위해 삽입한 데이터에 해당하는 이름을 제외한다.

ax.set_rlabel_position(45)    # 변경 정보 출력 각도
plt.yticks([20,40,60,80,100],['20','40','60','80','100'], color='red',size=8)    # 반경 정보
plt.ylim(0,100)    # 반경 범위

ax.plot(degree, kwanwoo, linewidth=1, linestyle='solid',color='orange', label = '관우')
ax.fill(degree, kwanwoo, 'orange', alpha = 0.2)

ax.plot(degree, kalyang, linewidth=1, linestyle='solid',color='orange', label = '제갈량')
ax.fill(degree, kalyang, 'blue', alpha = 0.1)

# 'best', 'upper right', 'upper left', 'lower left', 'lower right', 'right', 'center left', 'center right', 'lower center', 'upper center', 'center'
plt.legend(loc='lower right')
plt.show
```
