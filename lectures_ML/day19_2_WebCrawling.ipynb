웹사이트 주소를 보고 어디까지 기본 주소로 삼을 지를 판단.

#### urllib 날씨 정보 크롤링

```python
import urllib.request
url = "http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp"

values = {'stnId':'109'} # 딕셔너리 형태로 파라미터 정보 설정
params = urllib.parse.urlencode(values)    # 파라미터 인코딩. 영어 이외의 글자(숫자 포함)가 있는 경우 필요

url = url + '?' + params    # 주소와 인코딩된 파라미터의 결합
print('url=', url)

res=urllib.request.urlopen(url)
webpage=res.read().decode('utf-8')

print(webpage)
```

#### requests 날씨 정보 크롤링

```python
import requests

url = "http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp"

params = 'stnId=156'
url = url + '?' + params

print('url=', url)

webpage = requests.get(url)
print(webpage.text)
```

## Beautiful Soup

HTML 기호 제거의 필요성
- 수집된 웹 페이지에는 많은 HTML 기호가 있다.
이 기호들은 웹 브라우저가 적절한 양식으로 텍스트, 이미지 등을 표현할 때 필요하며, 웝 크롤링 시에도 원하는 부분을 추출하는 데에 도움이 된다.

- 그러나 일단 추출이 이루어진 다음에는 이 기호들이 더이상 필요가 없으므로 제거해준다.

### BeautifulSoup을 이용한 텍스트 추출
- 수집된 데이터는 특별한 이유가 없으면 HTML 기호를 제거한다.
- 정규식을 이용하여 원하는 부분만 추출할 수도 있으나 매번 상황에 맞게 코드를 작성하는 건 번거롭다.

- Beautiful Soup은 원하는 부분을 찾는 것 뿐 아니라 수집된 데이터에서 HTML 기호를 일괄 제거해주어 간단하게 텍스트를 추출할 수 있다.

- 아나콘다에 내장되어 있으나, conda, pip 등으로 최신 버전을 설치할 수도 있다.

### 주요 함수
- Beautiful Soup의 주요 함수로는 find(), find_all(), select()가 있다.
- 이 세 함수의 사용법에 대해 연습해본다.
- 웹 페이지를 직접 만들어 연습해본다.

#### find() 함수

```python
from bs4 import BeautifulSoup

webpage = '''
<html>
    <head>
        <title>오늘의 날씨</title>
    </head>
    <body>
        <p align='center'> 오늘의 날씨를 알려드리겠습니다.</p>
    </body>
</html>
'''

soup = BeautifulSoup(webpage,'html.parser')
soup.find('title')
```

HTML Parser 종류 : 
- html.parser : 내장된 parser. 적절한 속도. 추가 패키지 불요
- xml : 매우 빠르며, 구조가 깨져도 잘 동작함.
- html5lib : 매우 느리나, HTML 구조가 깨져도 매우 잘 동작함.

##### find() 함수의 한계
- find() 함수는 해당 요소가 여러 개 있을 때 처음 한 개만 출력한다.

```python
from bs4 import BeautifulSoup
webpage = '''
<html>
<head> <title> 오늘의 날씨 </title> </head>
<body>
<p align="center"> 오늘의 날씨를 알려드리겠습니다 </p>
<p align="right"> 내일의 날씨를 알려드리겠습니다 </p>
<p align="left"> 글피의 날씨를 알려드리겠습니다 </p>
</body>
</html>
'''
soup = BeautifulSoup(webpage, 'html.parser')
soup.find('p')
```

###### 속성을 이용한 접근
- 태그의 속성을 이용하면 구체적인 검색이 가능하다.
- 위의 경우 문단을 정렬하는 align 속성이 서로 다른 점을 이용할 수 있다.

```python
print(soup.find('p', align="center"))
print(soup.find('p', align="right"))
print(soup.find('p', align="left"))
```

#####  한 번에 여러 태그 검색
- find_all() 함수는 한 번에 여러 종류의 태그를 검색할 수도 있다.
    - 이때, 태그들은 괄호 등으로 감싸서 전달한다. (iterable 형태)
    - 동일한 태그를 여러개 전달하면 하나로 받아들인다. (집합처럼)

```python
soup.find_all(iter(('title', 'p', 'p')))
```

##### 텍스트 추출
- BeautifulSoup()의 함수들은 태그까지 같이 출력하므로, 출력된 결과에서 텍스트만 추출할 필요가 있다.
- .text를 이용해 텍스트만 추출한다.
    - .text 대신 get_text() 또는 getText() 도 동일한 결과를 반환한다.
    - 또는 .string 도 같은 역할을 한다. 다만 .string은 한 번에 한  문장밖에 가져오지 못한다
    
    

```python
html_all = soup.find_all(['title', 'p'])
print('original:', html_all[0])
print('.text :', html_all[0].text) 
```

##### 기상 예보 정보 추출
https://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp?stnId=156 의 내용을 살펴본다
- title
- wf

```python
import urllib.parse
from bs4 import BeautifulSoup
import urllib.request

url = 'http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp'

values = {'stnId':'109'}
params = urllib.parse.urlencode(values)

url = url + '?' + params
print('url :', url)

res = urllib.request.urlopen(url)
soup = BeautifulSoup(res, 'html.parser')

title = soup.find('title')
weather_forcast = soup.find('wf')

print("title :", title)
print('weather forcast :\n', weather_forcast)
```

```python
from bs4 import BeautifulSoup
import requests

url = 'http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp'
params = 'stnId=109'

url = url + '?' + params
print('url :', url)

res = requests.get(url)
soup = BeautifulSoup(res.text, 'html.parser')

title = soup.find('title').text
weather_forcast = soup.find('wf').text

print("title :", title)
print('weather forcast :\n', weather_forcast)

```
