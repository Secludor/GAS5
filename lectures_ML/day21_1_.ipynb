##### 네이버 환율 정보 추출
- 네이버 금융에서 환율 정보를 추출해 본다.
- https://finance.naver.com/marketindex/

````python
import urllib.request
from bs4 import BeautifulSoup
import urllib

url = 'https://finance.naver.com/marketindex/'

res = urllib.request.urlopen(url)
webpage = res.read().decode('euc-kr')

print(webpage)
````

````python
soup = BeautifulSoup(webpage, 'html.parser')
type(soup)
````

````python
spans = soup.select('div.head_info.point_dn>span.value')
# point_dn 은 수치가 직전대비 하락한 경우이다.

for span in spans:
    # span 요소의 부모 요소를 찾습니다
    parent_div = span.find_parent('div', class_='head_info point_dn')
        
    # 부모 요소 내에서 h3 요소를 찾습니다
    h3 = parent_div.find_previous_sibling('h3')
    
    if h3:
        print(h3.get_text(),' : ',span.text)
        print()
````

### 동적 크롤링

탭이 활성화된 경우에만 `resp-tab-active` 클래스를 가지는 요소를 크롤링하려면, 웹 페이지가 JavaScript를 사용하여 동적으로 콘텐츠를 로드하는 경우가 많기 때문에 단순히 `requests`와 `BeautifulSoup`만으로는 원하는 결과를 얻기 어려울 수 있습니다. 이러한 경우에는 `Selenium`을 사용하여 브라우저를 자동화하고, 원하는 탭을 클릭한 후에 `BeautifulSoup`으로 페이지 내용을 파싱하는 것이 일반적인 방법입니다.

##### Selenium을 사용한 예제

아래는 `Selenium`을 사용하여 특정 탭을 클릭하고, `BeautifulSoup`으로 해당 탭의 내용을 크롤링하는 예제 코드입니다:

1. **Selenium 설치**:
    ```sh
    pip install selenium
    ```

2. **웹 드라이버 설치** (예: ChromeDriver):
    - ChromeDriver 다운로드: https://sites.google.com/a/chromium.org/chromedriver/downloads
    - 다운로드한 드라이버를 시스템의 PATH에 추가하거나, 코드에서 직접 경로를 지정합니다.

3. **코드 작성**:

    ```python
    from selenium import webdriver
    from selenium.webdriver.common.by import By
    from selenium.webdriver.support.ui import WebDriverWait
    from selenium.webdriver.support import expected_conditions as EC
    from bs4 import BeautifulSoup

    # ChromeDriver 경로를 지정하세요
    driver_path = '/path/to/chromedriver'

    # 웹 드라이버 설정
    driver = webdriver.Chrome(driver_path)

    try:
        # 실제 URL을 사용하세요
        url = "https://example.com"
        driver.get(url)

        # 특정 탭 요소를 기다린 후 클릭 (예시: 탭 버튼의 ID나 클래스 이름을 사용)
        wait = WebDriverWait(driver, 10)
        tab_element = wait.until(EC.element_to_be_clickable((By.CLASS_NAME, 'tab-class-name')))
        tab_element.click()

        # 탭이 활성화되고 콘텐츠가 로드될 때까지 기다림
        wait.until(EC.presence_of_element_located((By.CLASS_NAME, 'resp-tab-active')))

        # 페이지 소스를 가져와서 BeautifulSoup으로 파싱
        soup = BeautifulSoup(driver.page_source, 'html.parser')

        # 클래스가 'resp-tab-active'인 div 요소 선택
        divs = soup.find_all('div', class_='resp-tab-active')

        for div in divs:
            print(div.prettify())  # div 요소의 내용을 출력 (prettify()로 보기 좋게 출력)

    finally:
        # 브라우저 종료
        driver.quit()
    ```

##### 설명

1. **Selenium 웹 드라이버 설정**:
    ```python
    driver = webdriver.Chrome(driver_path)
    ```

2. **웹 페이지 로드**:
    ```python
    driver.get(url)
    ```

3. **특정 탭 요소를 기다린 후 클릭**:
    ```python
    wait = WebDriverWait(driver, 10)
    tab_element = wait.until(EC.element_to_be_clickable((By.CLASS_NAME, 'tab-class-name')))
    tab_element.click()
    ```

4. **탭이 활성화되고 콘텐츠가 로드될 때까지 기다림**:
    ```python
    wait.until(EC.presence_of_element_located((By.CLASS_NAME, 'resp-tab-active')))
    ```

5. **페이지 소스를 가져와서 BeautifulSoup으로 파싱**:
    ```python
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    ```

6. **클래스가 'resp-tab-active'인 div 요소 선택 및 출력**:
    ```python
    divs = soup.find_all('div', class_='resp-tab-active')
    for div in divs:
        print(div.prettify())
    ```

##### 주의사항
- `ChromeDriver`의 경로를 정확히 지정해야 합니다.
- `WebDriverWait`과 `expected_conditions`를 사용하여 특정 요소가 로드될 때까지 기다리는 부분을 적절히 조정하세요.
- 실제 웹 페이지의 구조에 따라 `By.CLASS_NAME`, `By.ID`, `By.XPATH` 등을 사용하여 정확한 요소를 선택해야 합니다.

이 방법을 통해 JavaScript로 동적으로 로드되는 콘텐츠를 포함한 웹 페이지를 크롤링할 수 있습니다.

````python
import requests
from bs4 import BeautifulSoup

url = 'https://gj-aischool.or.kr/course/'

res = requests.get(url)

res
````

````python
soup = BeautifulSoup(res.text, 'html.parser')
````

````python
# "언어지능 특화과목" 텍스트가 포함된 요소 찾기
target_elements = soup.find_all(string="언어지능 특화과목")
````

````python
for element in target_elements:
    # 해당 텍스트가 포함된 부모 블록을 찾음 (예: div)
    parent_div = element.find_parent('div')
    if parent_div:
        print(parent_div.prettify())  # 부모 div 요소의 내용을 출력 (prettify()로 보기 좋게 출력)
````
